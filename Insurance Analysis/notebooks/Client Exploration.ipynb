# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

```python
# Let's load the csv and print the first 5 rows.
client_df = pd.read_csv('../src/data/insurance.csv')
print(client_df.shape)
client_df.head()
```

```python
# Let's check the types of each column.
client_df.dtypes
```

```python
# Look for missing values.
client_df.isnull().sum()
```

```python
# Look into their statistics.
client_df.describe()
```

```python
# Get the columns that are numeric.
for i in ['age', 'bmi','charges']:
    sns.displot(client_df[i], kde=True)
    plt.show()
    sns.boxplot(x=client_df[i])
    plt.show()
    print(f'The skew for {i} is: {client_df[i].skew()}')
```

```python
# Let's look at the categorical variables.
for i in client_df.select_dtypes(include=['object']).columns:
    sns.countplot(x=i, data=client_df)
    plt.show()
```

```python
# Make a distribution plot of charges, by smoker.
sns.displot(x='charges', data=client_df, hue='smoker', kde=True)
```

```python
# People who smoke have higher charges.
```

```python
# Distribution plot of bmi, by sex.
sns.displot(x='bmi', data=client_df, hue='sex', kde=True)
```

```python
# BMI are similar for both sex.
```

```python
sns.countplot(data=client_df, x='smoker', hue='sex')
```

```python
# Are similar for bot sexs the condition of smoker.
```

```python
# Distribution of bmi, for sex=female and children between 0 and 2.
sns.displot(data=client_df[(client_df['sex'] == 'female')& (client_df['children'] <= 2)], x='bmi', hue='children', kde=True)
```

```python
# Women with 2 children have a higher bmi, than 1 child, and 1 child higher than 0 children.
```
